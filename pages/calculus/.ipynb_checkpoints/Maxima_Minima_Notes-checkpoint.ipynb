```python
import matplotlib.pyplot as plt

x = range(-100, 100, 1)
y = []
for item in x:
    fx = pow(item, 2) - (4 * item) + 3
    y.append(fx)

plt.plot(x, y)
plt.show()
```

```python
y = []
for item in x:
    fx = 3 + (2 * item) - (pow(item, 2))
    y.append(fx)

plt.plot(x, y)
plt.show()    
```

```python
import numpy as np

x = np.arange(0, 3, 0.01)
print(len(x))
y = []
for item in x:
    fx = (item - 1) * (item - 2) * (item - 3)
    y.append(fx)

plt.grid(True)
plt.plot(x, y)
plt.show()
```
