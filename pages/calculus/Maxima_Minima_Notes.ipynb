```python
import matplotlib.pyplot as plt

x = range(-100, 100, 1)
y = []
for item in x:
    fx = pow(item, 2) - (4 * item) + 3
    y.append(fx)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
x = np.arange(-5, 5, 0.01)

y = []
for item in x:
    fx = pow(item, 2)
    y.append(fx)

plt.title('Graph of y= x²')

#plt axes
origin = [0, 0]
axis_1 = [25, -5]
axis_2 = [-5, 5]
plt.plot(origin, axis_1, color="black", linestyle='--')
plt.plot(axis_2, origin, color="black", linestyle='--')

# plot the curve
plt.plot(x, y)

# Plotting the max/min point
plt.scatter(0, 0, color='red', marker='o')

# Adding coordinates in text above the point
plt.text(0, -2, 'minimum point', ha='center', color="red")
plt.text(0, -1, '(0, 0)', ha='center', color="red")


plt.grid(True)


# Save the figure with a transparent background and specified size
plt.savefig("minimum_point.png", transparent=True, dpi=78)

plt.show()    
```

```python
#### import numpy as np
import matplotlib.pyplot as plt


#plt axes
origin = [0, 0]
axis_1 = [6, -6]
axis_2 = [0, 6]
plt.plot(origin, axis_1, color="black", linestyle='--') # y axis
plt.plot(axis_2, origin, color="black", linestyle='--') # x axis

# plt curve
x = np.arange(0, 4, 0.01)
print(len(x))
y = []
for item in x:
    fx = (item - 1) * (item - 2) * (item - 3)
    y.append(fx)
plt.plot(x, y, color="blue")

# plt gradient function
x = np.arange(0.5, 3.5, 0.01)
dx = []
for item in x:
    fx = (3 * (pow(item, 2))) - (12 * item) + 11
    dx.append(fx)
plt.plot(x, dx, color="red", linestyle='--')

# plt second order differential
x = np.arange(1, 3, 0.01)
d2x = []
for item in x:
    fx = (6 * item) - 12
    d2x.append(fx)
plt.plot(x, d2x, color="green", linestyle='--')

# plt maximum, minimum and points of inflexion
x = [1.42, 2.58]
y = [0.385, -0.385]
plt.scatter(x, y)
plt.text(x[0], y[0] + 0.2, 'maximum point', ha='center', color="blue")
plt.text(x[1], y[1] - 0.35, 'minimum point', ha='center', color="blue")


plt.text(4, 5, 'y = (x - 1)(x - 2)(x - 3)', color="blue")
plt.text(0.75, 5, 'y = 3x² - 12x + 11', color="red")
plt.text(1.5, -4, 'y = 6x - 12', color="green")

plt.grid(True)

# Save the figure with a transparent background and specified size
plt.savefig("inflexion_point.png", transparent=True, dpi=100)

plt.show()
```
