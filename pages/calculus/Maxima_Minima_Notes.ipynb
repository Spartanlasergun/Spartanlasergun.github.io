```python
import matplotlib.pyplot as plt

x = range(-100, 100, 1)
y = []
for item in x:
    fx = pow(item, 2) - (4 * item) + 3
    y.append(fx)

plt.plot(x, y)
plt.show()
```

```python
y = []
for item in x:
    fx = -1 * pow(item, 2)
    y.append(fx)

plt.title('Graph of y= -x²')
plt.plot(x, y)
plt.grid(True)


# Save the figure with a transparent background and specified size
plt.savefig("maximum_point.png", transparent=True, dpi=72)

plt.show()    
```

```python
import numpy as np

x = np.arange(0, 3, 0.01)
print(len(x))
y = []
for item in x:
    fx = (item - 1) * (item - 2) * (item - 3)
    y.append(fx)

plt.grid(True)
plt.plot(x, y)
plt.show()
```

```python
x = range(-100, 100, 1)
y = []
for item in x:
    fx = (pow(item, 2)) - (2 * item)
    y.append(fx)

plt.plot(x, y)
plt.show()
```

```python
x = range(-100, 100, 1)
y = []
for item in x:
    fx = (3 * item) - (pow(item, 2))
    y.append(fx)

plt.plot(x, y)
plt.show()
```
